```python
from glob import glob
import pandas as pd
get_datafile = glob('*.csv')
print(get_datafile)
```

```python
for i in get_datafile:
  df_train = pd.read_csv(i)
  text=[]
  label=[]
  for index,row in df_train.iterrows():
    a_split = row['text;label'].split(';')
    text.append(a_split[0])
    label.append(a_split[1])

  final = {'text': text, 'label': label} 
  df = pd.DataFrame(final)
  df.to_csv('Data_'+i+'.csv')
```
