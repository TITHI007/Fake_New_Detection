```python
from glob import glob
lis = glob('*.csv')
print(lis)
```

```python
lis=['KAGGLE_DATASET_Jruvika.csv', 'ISOT_FAKE_NEWS_DATASET.csv', 'UTK_FAKE_NEWS_DATASET.csv', 'G_McIntire_FAKE_NEWS_DATASET_.csv', 'COVID-19_Tweets_FAKE_NEWS_DATASET.csv','LIAR_FAKE_NEWS_DATASET.csv' ]
```

```python
lis[2:]
```

```python
rows=[]
for i in lis[2:] :
  # read training data
  df_train = pd.read_csv(i)
  
  # prepare data
  n_divisions = 5
  print(n_divisions)

  X_train = df_train['text']
  y_train = df_train['label']

  # extract features
  # build a dictionary of words for the training
  corpus = [str (i) for i in X_train]
  # extract the bag of words and frequencies for the training
  vectorizer = CountVectorizer(stop_words="english", min_df=0.01)
  vect_for_df_svm = vectorizer.fit_transform(corpus)
  X_train = vect_for_df_svm.toarray()

  # build a model to use with the new data
  clf = BaggingClassifier(base_estimator=SVC(kernel='linear', C=1),
                        n_estimators=n_divisions, random_state=0).fit(X_train, y_train)
      # , n_jobs=-1
  j_compare=[]
  j_compare.append(i)
  for j in lis:

    # read test data
    df_test = pd.read_csv(j)

    X_test = df_test['text']
    y_test = df_test['label']
    
    X_test=[str (i) for i in X_test]
    X_test = vectorizer.transform(X_test).toarray()

    # test the model in the unseen data                        
    y_predicted = clf.predict(X_test)

    # compute performance
    performances = np.asarray(precision_recall_fscore_support(y_test,y_predicted, average='macro'))
    j_compare.append(performances[2])
    # save results
    print(performances)
    print(i+j+"completed")
  rows.append(j_compare)
    
```

```python
import csv
fields=['','KAGGLE_DATASET_Jruvika.csv', 'ISOT_FAKE_NEWS_DATASET.csv', 'UTK_FAKE_NEWS_DATASET.csv', 'G_McIntire_FAKE_NEWS_DATASET_.csv', 'COVID-19_Tweets_FAKE_NEWS_DATASET.csv', 'LIAR_FAKE_NEWS_DATASET.csv']

with open('../RESULT/SVM_RESULT2.csv', 'w') as f:
      
    # using csv.writer method from CSV package
    write = csv.writer(f)
      
    write.writerow(fields)
    write.writerows(rows)
```
