```python
import pandas as pd
```

```python
from glob import glob
lis = glob('*.csv')
print(lis)
TRUE_FALSE=[]
for i in lis:
  df_train = pd.read_csv(i)
  count_True=0
  count_False=0
  for index,row in df_train.iterrows():
    if row['label'] == 0:
      count_True=count_True+1
    else:
      count_False=count_False+1
  TRUE_FALSE.append([i,count_True,count_False])
  print(TRUE_FALSE)
```

```python
import csv
header=["DATASET","Real_news","Fake_news"]
with open('DATA_Division.csv', 'w') as f:
      
    # using csv.writer method from CSV package
    write = csv.writer(f)
      
    write.writerow(header)
    write.writerows(TRUE_FALSE)
```
